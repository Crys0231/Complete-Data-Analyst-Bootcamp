## Libraries

```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set_theme() # this command sets the seaborn chart style as the default

from matplotlib.ticker import PercentFormatter #converts values into percentage format
```

## Bar Chart

```python
df_used_cars = pd.read_csv("S34 - Data Visualization/L1/bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize=(10,6))
plt.bar(x = df_used_cars['Brand'], height= df_used_cars['Cars Listings'], color = "#028090")
plt.xticks(rotation = 45, fontsize= 13)
plt.title("Cars Listing by Brand", fontsize = 16, fontweight = 'bold')
plt.ylabel("Number of Listings", fontsize = 13)
plt.show()
plt.savefig("Used_Cars_Bar.png")
```

## Pie Chart

```python
df_fuel_engine_type = pd.read_csv('pie_chart_data.csv')
```

```python
df_fuel_engine_type
```

```python
sns.color_palette('colorblind')
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_fuel_engine_type['Number of Cars'], 
        labels = df_fuel_engine_type['Engine Fuel Type'].values, 
        autopct='%.2f%%', 
        textprops= {'size' : 'x-large', 
                    'rotation': 'horizontal'})

plt.legend()
plt.title('Cars by Engine Fuel Type', fontsize = 16, fontweight = 'bold')
plt.show()
```

## Stacked Area Chart

```python
df_fuel_engine_types = pd.read_csv('stacked_area_chart_data.csv')
```

```python
df_fuel_engine_types.head()
```

```python
colors = ['#e5383b', '#a4161a', '#660708']
labels = ['Diesel', 'Petrol', 'Gas']
plt.figure(figsize = (12,6))

plt.stackplot(df_fuel_engine_types['Year'], df_fuel_engine_types['Diesel'], df_fuel_engine_types['Petrol'], df_fuel_engine_types['Gas'], colors=colors, labels=labels)
plt.legend(labels = labels, loc = 'upper left')
plt.ylabel('Number of Cars', fontsize=13)
plt.title('Popularity of Engine Fuel Types (1982 - 2016)', fontsize = 14, weight = 'bold')
plt.xticks(df_fuel_engine_types['Year'], rotation = 45)
sns.despine()
plt.show()
```

## Line Chart

```python
df_ftse = pd.read_csv('line_chart_data.csv')
```

```python
df_ftse
```

```python
df_ftse['new_date'] = pd.to_datetime(df_ftse['Date'])
```

```python
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize= (20,8))
plt.plot(df_ftse['new_date'], df_ftse['GSPC500'])
plt.plot(df_ftse['new_date'], df_ftse['FTSE100'])
plt.title('S&P vc FTSE Returns (2000 - 2010)', fontsize = 14, fontweight = 'bold')
plt.ylabel('Returns')
plt.xlabel('Date')
plt.legend(labels = labels, fontsize ='large' )
plt.show()
```

```python
df_ftse_half = df_ftse[(df_ftse.new_date >= '2008-07-01') & (df_ftse.new_date <= '2008-12-31')]
```

```python
df_ftse_half
```

```python
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize= (20,8))
plt.plot(df_ftse_half['new_date'], df_ftse_half['GSPC500'], color = '#4f6272')
plt.plot(df_ftse_half['new_date'], df_ftse_half['FTSE100'], color = '#dd7596')
plt.title('S&P vc FTSE Returns (H2 2008)', fontsize = 14, fontweight = 'bold')
plt.ylabel('Returns')
plt.xlabel('Date')
plt.legend(labels = labels, fontsize ='large' )
plt.show()
```

## Histogram

```python
df_real_state = pd.read_csv("histogram_data.csv")
```

```python
sns.set_style('dark')
plt.figure(figsize = (8,6))
plt.hist(df_real_state['Price'], bins =8, color = "#108A99")
plt.title("Distribution of Real Estate Prices", fontsize = 14, weight = "bold")
plt.xlabel("Price in (000' $)")
plt.ylabel("Number of Properties")
sns.despine()
plt.show()
```

## Scatter Plot

```python
df_scatter = pd.read_csv("scatter_data.csv")
df_scatter
```

```python
plt.figure(figsize= (12,8))
sns.set_style('dark')
scatter = plt.scatter(df_scatter['Price'], df_scatter['Area (ft.)'],alpha = 0.6, c = df_scatter['Building Type'], cmap = 'viridis')
plt.legend(*scatter.legend_elements(), loc = 'upper left', title = 'Building Type')
plt.title("Relationship between Area and Price of California Real Estate", fontsize = 14, weight = 'bold' )
plt.xlabel("Area (sq. ft)", weight = 'bold')
plt.ylabel("Price (000' of $)")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = df_scatter['Price'], 
                y = df_scatter['Area (ft.)'], 
                hue = df_scatter['Building Type'], 
                palette= ['black', 'darkblue','purple', 'pink', 'white' ], 
                s = 100)
plt.title("Relationship between Area and Price of California Real Estate", fontsize = 14, weight = 'bold' )
plt.xlabel("Area (sq. ft)", weight = 'bold')
plt.ylabel("Price (000' of $)")
plt.show()
```

## Regression Plot

```python
df_scatter_ii = pd.read_csv('scatter_plot_ii.csv')
```

```python
sns.set(rc = {'figure.figsize': (9,6)})
plt.figure(figsize = (10,8))
sns.regplot(x = 'Budget',
            y = 'Sales',
            data = df_scatter_ii,
            scatter_kws = {'color':  'k'},
            line_kws= {'color': 'red'})
plt.xlabel("Ad Expenditure in (000' of Units)")
plt.ylabel("sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 15, weight = 'bold')
plt.show()
```

```python
sns.lmplot(x = 'Budget', 
           y = 'Sales', 
           data = df_scatter_ii, 
           height=8, 
           scatter_kws = {'color':  'k'},
           line_kws= {'color': 'red'})

# axis labels

plt.xlabel("Ad Expenditure in (000' of Units)")
plt.ylabel("sales in (000's Units)")

# title
plt.title("Effect of Ad Expenditure on Sales", fontsize = 15, weight = 'bold')

plt.show()
```

## Bar and Line Chart

```python
from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv('bar_line_chart_data.csv')
```

```python
fig, ax = plt.subplots(figsize = (10,7))
ax.bar(df_kdnuggets['Year'],
       df_kdnuggets['Participants'],
       color = 'grey')

ax.set_ylabel("Number of Participants", weight = 'bold')
ax.tick_params(axis = 'y', width = 2, labelsize = 'large')


ax1 = ax.twinx()
ax1.set_ylim(0,1)

#puts the max percentage to 100
ax1.yaxis.set_major_formatter(PercentFormatter(xmax= 1.0))
ax1.plot(df_kdnuggets['Year'], df_kdnuggets['Python Users'], color = '#dd1c1a', marker = 'D')
ax1.set_ylabel("Python Users", color = '#dd1c1a', weight = 'bold')
ax1.tick_params(axis = 'y', colors = '#dd1c1a', width = 2, labelsize = 'large')

ax.set_title("KD Nuggets Survey Python Users (2012 - 2019)", fontsize = 14, weight = 'bold')
```
